##**SERIES**

Pandas Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.). The axis labels are collectively called index. Pandas Series is nothing but a column in an excel sheet.
Labels need not be unique but must be a hashable type. The object supports both integer and label-based indexing and provides a host of methods for performing operations involving the index.

###**1-Series from a LIST**

```
# Create a pandas data series from a list:

L1 = ['Honda', 'Toyota','Ford','Tesla','Lucid','Rivian']

pseries_1 = pd.Series(data = L1)
pseries_1
```

```
#create a pandas data series with index
car_index = ['car1','car2','car3','ev1','ev2','ev3']

pseries_2 = pd.Series(data = L1,index = car_index)
pseries_2
```

###**17-Filter a Dataframe**

###**24-Concatenate Dataframes**

```
# Using CONCAT() Method
ev_cars = pd.DataFrame([['Tesla', 'EV'], ['Lucid', 'EV']],
                   columns=['Car_Name', 'Car_Type'])

ic_cars = pd.DataFrame([['Honda', 'IC'], ['Toyota', 'IC']],
                   columns=['Car_Name', 'Car_Type'])

# Now merge the 2 dataframes
all_cars = pd.concat([ev_cars, ic_cars],ignore_index=True)
all_cars
```

```
# Using APPEND Method
all_cars = ev_cars.append(ic_cars,ignore_index = True)
all_cars
```

```
# Using MERGE Method
all_cars = pd.DataFrame([['Honda', 25000], ['Toyota', 27000],['Nissan', 45000], ['BMW', 57000],['Tesla', 50000], ['Lucid', 100000]],
                   columns=['Car_Name', 'Price'])
mileage_cars = pd.DataFrame([['Honda','IC', 22], ['Toyota','IC',25],['Nissan','Hybrid', 80], ['BMW','Hybrid', 100],['Tesla','EV', 350],['Lucid','EV', 340]],
                             columns=['Car_Name','Car_Type', 'Mileage'])

#Merge the 2 dataframe on Car_Name
cars_merged = pd.merge(all_cars, mileage_cars, on = 'Car_Name')
cars_merged
```

###**25-Groupby & Apply Functions**

* Use the groupby() method.
* After the groupby then you can apply an aggregate function
* groupby splits the dataframe based on the column condition. For example in this case the dataframe is split based on the passenger classes which are 1,2 and 3.
* Then it applies the aggregate function, in this case, apply sum to all the integer and float data types column.
* Then it displays the data for that grouping Id like 1,2 and 3.
* You can also groupby multiple columns. Pass the column names by a list.

```
titanic_df = pd.read_csv("titanic.csv")
titanic_df.groupby("Pclass").sum()
```

```
#You can also group by multiple columns. here groupby sex and passenger class. Then apply sum
titanic_df.groupby(['Sex','Pclass'])[['Fare']].sum()
```

```
#Use the get_group() method.
titanic_df = pd.read_csv('titanic.csv')
gender_df = titanic_df.groupby('Sex')
gender_df.get_group('Male')
```

###**26-Apply Agg Method after Groupby**

###**27-Pandas Time Data**

Date and time features find importance in data science problems spanning industries from sales, marketing, and finance to HR, e-commerce, retail, and many more. Predicting how the stock markets will behave tomorrow, how many products will be sold in the upcoming week, when is the best time to launch a new product, how long before a position at the 

```
# DATE
from datetime import date

d1 = date.today()
print(d1)
# day
print('Day :',d1.day)
# month
print('Month :',d1.month)
# year
print('Year :',d1.year)
```

```
# TIME
from datetime import time

t1 = time(13,20,13,40)
# hour
print('Hour :',t1.hour)
# minute
print('Minute :',t1.minute)
# second
print('Second :',t1.second)
# microsecond
print('Microsecond :',t1.microsecond)
```

```
# DATE-TIME-NOW
from datetime import datetime
d1 = datetime.now()
print(d1)
print(type(d1))
```

```
d1 = datetime.now()
# week starts from 0
print(d1.weekday()) # output 3 for Thurday
# week starts with 1
print(d1.isoweekday()) # output 4 in ISO format
# retuns year, week, month
print(d1.isocalendar())
print('Week :',d1.isocalendar()[1])
```

```
import calendar
d1 = datetime.now()
# leap year or not
calendar.isleap(d1.year) # Output True
```

```
print(calendar.calendar(2020))
```

```
# strptime
date = '22 April, 2020 13:20:13'
d1 = datetime.strptime(date,'%d %B, %Y %H:%M:%S')
print(d1)
```

```
# strftime
d1 = datetime.now()
print('Datetime object :',d1)
new_date = d1.strftime('%d/%m/%Y %H:%M')
print('Formatted date :',new_date)
```

```
# timedelta : duration between dates
d1 = datetime(2020,4,23,11,13,10)
d2 = datetime(2021,4,23,12,13,10)
duration = d2-d1
print(type(duration))
duration
```
